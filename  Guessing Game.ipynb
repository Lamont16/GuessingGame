```python
import random

name = input("What is your name? ")
print("Good Luck ! ", name)

words = ['tennis', 'mickey mouse', 'disney world', 'coding temple',
        'python', 'atlanta', 'phenomenal', 'princess',
        'goofy', 'falcons', 'hawks', 'braves']

word = random.choice(words)


print("Guess the letters")

guesses = ''


turns = 7


while turns > 0:
    failed = 0
    for letter in word:
        if letter in guesses:
            print(letter, end=" ")
        else:
            print("_")
            print(end=" ")
            failed += 1
    if failed == 0:
        print("You Win")
        print("The word is: ", word)
        break
    guess = input("guess a letter:")
    guesses += guess
if guess not in word:
            turns -= 1
            print("Sorry Wrong Answer")
            print("You have", + turns, 'more guesses')


if turns == 0:
        print("You Loose. Play again? ")
```
